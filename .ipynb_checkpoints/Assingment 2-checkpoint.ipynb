This is the environment and other functions that are used in it and the agent.

```python

class Gold(Thing):
    pass

class Trap(Thing):
    pass



def onedge(ap):
    #if (p.location[1] == 4 and p.direction == "right") or (p.location[1] == 0 and p.direction == "left") or (p.location[0] == 4 and p.direction == "down") or (p.location[0] == 0 and p.direction == "up"):
    #    return True
    if (ap[0] >= 5 or ap[1] >= 5 or ap[1] <= -1 or ap[0] <= -1):
        return True
    else:
        return False

def manhattan(loc1,loc2):
    return abs(loc1[0]-loc2[0])+abs(loc1[1]-loc2[1])

def getgoal(perps,p):
    closestdistance = 99
    for k in perps:
        if isinstance(k[0], Gold):
            if closestdistance == 99 or (manhattan(p[1],k[1]) <= closestdistance):
                closestdistance = manhattan(p[1],k[1])
                goal = k[1]
            #print("current closest distance = "+str(goal))
    return goal

def getgoalanddist(perps,p):
    closestdistance = 99
    for k in perps:
        if isinstance(k[0], Gold):
            if closestdistance == 99 or (manhattan(p[1],k[1]) <= closestdistance):
                closestdistance = manhattan(p[1],k[1])
                goal = k[1]
            #print("current closest distance = "+str(goal))
    return goal,closestdistance


def amITrapped(perps,p):
    for k in perps:
        if isinstance(k[0], Trap):
            if (manhattan(p[1],k[1]) == 0):
                return True
    return False


def getadvanceposition(p):
    if p[0].direction=="right":
        return [p[1][0],p[1][1]+1]
    elif p[0].direction == "left":
        return [p[1][0],p[1][1]-1]
    elif p[0].direction == "up":
        return [p[1][0]-1,p[1][1]]
    elif p[0].direction == "down":
        return [p[1][0]+1,p[1][1]]
    
    
    
def istheregold(perps):
    for p in perps:
        if isinstance(p[0],Gold):
            return True
    return False
    


class Grid(Environment):
    def percept(self, agent):
        things = []
        '''prints & return a list of things that are in our agent's location'''
        for x in range(0,5):
            for y in range(0,5):
                if(len(self.list_things_at([x,y]))!= 0):
                    for thing in self.list_things_at([x,y]):                        
                        things.append([thing,[x,y]])        
        #print(things)
        return things
    def execute_action(self, agent, action):
        chargold = ""
        chartrap = ""
        charhero = "R"
        '''changes the state of the environment based on what the agent does.'''
        if action == "advance":
            itemsGold = self.list_things_at(getadvanceposition([agent,agent.location]), tclass=Gold)
            itemsTraps = self.list_things_at(getadvanceposition([agent,agent.location]), tclass=Trap)
            if len(itemsGold) == 0 and len(itemsTraps) == 0:
                agent.advance()
            elif len(itemsGold) != 0 and len(itemsTraps) == 0:
                if agent.advanceAndGetAffected([itemsGold[0]]):
                    self.delete_thing(itemsGold[0])                   
            elif len(itemsGold) == 0 and len(itemsTraps) != 0:
                if agent.advanceAndGetAffected([itemsTraps[0]]):
                    self.delete_thing(itemsTraps[0])                    
            else:
                if agent.advanceAndGetAffected([itemsGold[0],itemsTraps[0]]):
                    self.delete_thing(itemsGold[0])
                    self.delete_thing(itemsTraps[0])
        elif action == "turn":
            itemsGold = self.list_things_at(agent.location, tclass=Gold)
            itemsTraps = self.list_things_at(agent.location, tclass=Trap)
            if len(itemsGold) == 0 and len(itemsTraps) == 0:
                agent.turn()
            elif len(itemsGold) != 0 and len(itemsTraps) == 0:
                if agent.turnAndGetAffected([itemsGold[0]]):
                    self.delete_thing(itemsGold[0])                   
            elif len(itemsGold) == 0 and len(itemsTraps) != 0:
                if agent.turnAndGetAffected([itemsTraps[0]]):
                    self.delete_thing(itemsTraps[0])                    
            else:
                if agent.turnAndGetAffected([itemsGold[0],itemsTraps[0]]):
                    self.delete_thing(itemsGold[0])
                    self.delete_thing(itemsTraps[0])                     
        elif action == "stay":
            itemsGold = self.list_things_at(agent.location, tclass=Gold)
            itemsTraps = self.list_things_at(agent.location, tclass=Trap)
            if len(itemsGold) == 0 and len(itemsTraps) == 0:
                agent.stay()
            elif len(itemsGold) != 0 and len(itemsTraps) == 0:
                if agent.stayAndGetAffected([itemsGold[0]]):
                    self.delete_thing(itemsGold[0])                   
            elif len(itemsGold) == 0 and len(itemsTraps) != 0:
                if agent.stayAndGetAffected([itemsTraps[0]]):
                    self.delete_thing(itemsTraps[0])                    
            else:
                if agent.stayAndGetAffected([itemsGold[0],itemsTraps[0]]):
                    self.delete_thing(itemsGold[0])
                    self.delete_thing(itemsTraps[0])                
        for x in range(0,5):
            for y in range(0,5):
                if(agent.location == [x,y]):
                    if(agent.direction == "right"):
                        charhero = 'R'
                    elif(agent.direction == "left"):
                        charhero = 'L'
                    elif(agent.direction == "up"):
                        charhero = 'U'
                    elif(agent.direction == "down"):
                        charhero = "D"
                else:
                    charhero = '-'
                    
                thingsgold = self.list_things_at([x,y], tclass=Gold)
                thingstrap = self.list_things_at([x,y],tclass=Trap)
                if len(thingsgold) != 0 or len(thingstrap) != 0:
                    if(len(thingsgold) == 0):
                        chargold = "-"
                    else:
                        chargold = str(len(thingsgold))
                    if(len(thingstrap) == 0):
                        chartrap = "-"
                    else: 
                        chartrap = str(len(thingstrap))
                    print("("+charhero+","+chargold+","+chartrap+") ", end = '')
                else:
                    print("("+charhero+",-,-) ", end = '')
            print("\n")
        print(str(agent.points))
            
            
    def add_thing(self, thing, location=None):
        """Add a thing to the environment, setting its location. For
        convenience, if thing is an agent program we make a new agent
        for it. (Shouldn't need to override this."""
        if not isinstance(thing, Thing):
            thing = Agent(thing)
        assert thing not in self.things, "Don't add the same thing twice"
        thing.location = location if location is not None else self.default_location(thing)
        self.things.append(thing)
        if isinstance(thing, Agent):
            thing.performance = 0
            self.agents.append(thing)
    
    def is_done(self):
        '''By default, we're done when we can't find a live agent, 
        but to prevent killing our cute dog, we will or it with when there is no more food or water'''
        #no_edibles = not any(isinstance(thing, Food) or isinstance(thing, Water) for thing in self.things)
        dead_agents = not any(agent.is_alive() for agent in self.agents)
        return dead_agents
```

This is the Agent

```python

class Hero(Agent):
    location = [1,1]
    points = 100
    direction = "right"
    
    def advance(self):
        if(self.direction == "right"):
            self.location[1]+=1
        elif(self.direction == "left"):
            self.location[1]-=1
        elif(self.direction== "up"):
            self.location[0] -=1
        elif(self.direction== "down"):
            self.location[0] += 1
        self.points -= 1
    def advanceAndGetAffected(self,lista):
        if(self.direction == "right"):
            self.location[1]+=1
        elif(self.direction == "left"):
            self.location[1]-=1
        elif(self.direction== "up"):
            self.location[0] -=1
        elif(self.direction== "down"):
            self.location[0] += 1
        self.points -= 1
        for x in lista:
            if isinstance(x,Gold):
                print("I found a coint at:  {}.!".format(self.location))
                self.points += 10
            if isinstance(x,Trap):
                print("I got trapped at:  {}.!".format(self.location))
                self.points -= 5
        return True
    def turn(self):
        if(self.direction == "right"):
            self.direction = "down"
        elif(self.direction == "left"):
            self.direction = "up"
        elif(self.direction== "up"):
            self.direction ="right"
        elif(self.direction== "down"):
            self.direction = "left"
        self.points -= 1
    def turnAndGetAffected(self,lista):
        if(self.direction == "right"):
            self.direction = "down"
        elif(self.direction == "left"):
            self.direction = "up"
        elif(self.direction== "up"):
            self.direction ="right"
        elif(self.direction== "down"):
            self.direction = "left"
        self.points -= 1
        for x in lista:
            if isinstance(x,Gold):
                print("I found a coint at:  {}.!".format(self.location))
                self.points += 10
            if isinstance(x,Trap):
                print("I got trapped at:  {}.!".format(self.location))
                self.points -= 5
        return True
    def stay(self):
        print("Stay")
    def stayAndGetAffected(self,lista):
        print("Stay")
        for x in lista:
            if isinstance(x,Gold):
                print("I found a coint at:  {}.!".format(self.location))
                self.points += 10
            if isinstance(x,Trap):
                print("I got trapped at:  {}.!".format(self.location))
                self.points -= 5
        return True        


    
def program2(percepts):
    '''Returns an action based on it's percepts'''
    if istheregold(percepts) == False:
        return 'stay'
    else:
        for p in percepts:
            if isinstance(p[0], Hero):
                goal = getgoalanddist(percepts,p) #Closest gold piece
                ap = getadvanceposition(p) #Next position in current trajectory
                print("My goal is: "+str(goal[0])+" and my next move is: "+str(ap))
                if onedge(ap):
                    return 'turn'
                else:
                    if manhattan(goal[0],ap) < goal[1]:
                        print("I have to move!")
                        return 'advance'
                    elif manhattan(goal[0],p[1]) == 0:
                        return 'stay'
                    elif amITrapped(percepts,p):
                        return 'advance'
                    else: 
                        return 'turn'
    return 'advance'
```

This is the test

```python
grid = Grid()
hero = Hero(program2)
goldpiece = Gold()
goldpiece2 = Gold()
goldpiece3 = Gold()
goldpiece4 = Gold()
goldpíece5 = Gold()
trap1 = Trap()
trap2 = Trap()
trap = Trap()
grid.add_thing(hero, [0,0])
grid.add_thing(goldpiece, [2,2])
grid.add_thing(trap, [3,3])
grid.add_thing(goldpiece2, [2,2])
grid.add_thing(goldpíece5, [4,4])
grid.add_thing(goldpiece4, [3,1])
grid.add_thing(goldpiece3, [3,3])
grid.add_thing(trap1, [2,1])
grid.add_thing(trap2, [2,1])
grid.run(23)
```
